2.) REGION:

```python
# Filter the DataFrame for the 'Central' region
central_region_df = df[df['Region'] == 'Central'].copy()

# Group by Item in the Central region and calculate the sum of 'Total' sales
central_item_sales = central_region_df.groupby('Item')['Total'].sum().reset_index()

# Sort the items by total sales for better visualization (optional)
central_item_sales = central_item_sales.sort_values('Total', ascending=False)

# Create a line chart for the Central region
plt.figure(figsize=(10, 6))
plt.plot(central_item_sales['Item'], central_item_sales['Total'], marker='o', linestyle='-')

# Add labels and title
plt.title('Total Sales by Item in Central Region (Line Chart)')
plt.xlabel('Item')
plt.ylabel('Total Sales')

# Rotate x-axis labels if necessary for readability
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

# Show the chart
plt.show()
```
