School supplies Store Sales Analysis

1.) DATE:

```python
df_2021_12_29 = df[df['OrderDate'] == '2021-12-29']
display(df_2021_12_29)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

df_2021_12_29 = df[df['OrderDate'] == '2021-12-29'].copy()

plt.figure(figsize=(8, 6))

x_values = np.arange(len(df_2021_12_29))
plt.bar(x_values, df_2021_12_29['Total'], color='skyblue')

plt.title('Sales on 2021-12-29')
plt.xlabel('Order Date')
plt.ylabel('Total Sales')

plt.xticks(x_values, df_2021_12_29['OrderDate'].dt.strftime('%Y-%m-%d'), rotation=45, ha='right')
plt.tight_layout()

plt.show()
```

INTERPRETATION: <br> The bar chart shows that two separate sales happened on 2021-12-29. Both bars are the same height, which means each sale had the same total amount. The x-axis is for the date, and the y-axis shows the sales value.

2.) REGION:

```python
df_cregion = df[df['Region'] == 'Central']
display(df_cregion)
```

```python
central_region_df = df[df['Region'] == 'Central'].copy()

central_item_sales = central_region_df.groupby('Item')['Total'].sum().reset_index()

central_item_sales = central_item_sales.sort_values('Total', ascending=False)

plt.figure(figsize=(10, 6))
plt.plot(central_item_sales['Item'], central_item_sales['Total'], marker='o', linestyle='-')


plt.title('Total Sales by Item in Central Region (Line Chart)')
plt.xlabel('Item')
plt.ylabel('Total Sales')

plt.xticks(rotation=45, ha='right')
plt.tight_layout()

plt.show()
```

INTERPRETATION: <br>
The line chart shows the total sales for each item in the Central region. It shows on the chart that  the Binder had the highest sales, followed by Pen Set, then Pencil, Desk, and Pen, which had the lowest sales among these items. Using this line chart, makes it easy to see which items are selling the most in the Central region.

3.) REPRESENTATIVE:

```python
df_jardine = df[df['Rep'] == 'Jardine']
display(df_jardine)
```

```python
import matplotlib.pyplot as plt

jardine_item_units = df_jardine.groupby('Item')['Units'].sum()

plt.figure(figsize=(8, 8))
plt.pie(jardine_item_units, labels=jardine_item_units.index, autopct='%1.1f%%', startangle=140)

plt.title('Units Sold by Item for Jardine')

plt.axis('equal')

plt.show()
```

INTERPRETATION: <br>
The pie chart allows us to see which items were sold the most  in percentage. Pencil takes up the biggest slice at 44.8%, which means Jardine sold the most pencils. Binder is next at 37.4%, and Pen Set makes up the remaining 17.8% of units sold.
